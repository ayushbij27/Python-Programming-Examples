```python
import pandas as pd
import numpy as np
import os
import sys
filename="C:\Users\Ayush\Desktop\Python Exercises\/nycflights_med.csv"
data_frame=pd.read_csv(filename,sep=',')
# with open(filename,'r') as f:
# 	data_file=csv.DictReader(f)
# 	for i in data_file:
# 		print i
```

```python

data_frame['date'] = pd.to_datetime(data_frame['date'])
```

```python
# Calculating week_number and weekday
data_frame['Week_Number'] = data_frame['date'].dt.week
data_frame['weekday'] = data_frame['date'].dt.dayofweek
```

```python
# Counting flights per day of the week
data_frame['flights per day'] = data_frame.groupby(['date'])['weekday'].transform('count')
```

```python
data_frame
```

```python
# Counting number of delayed flights on weekend
data_frame[(data_frame['arr_delay']==0) & (data_frame['weekday']>4)].count()['flights per day']
```

```python
# Counting number of delayed flights on weekday
data_frame[(data_frame['arr_delay']==0) & (data_frame['weekday']<5)].count()['flights per day']
```

```python
# Classiifying distance into distance _class
data_frame['distance_class']= np.where(data_frame['distance']<1500, 'short', np.where((data_frame['distance']>1500) & (data_frame['distance']< 3000),'medium', 'large'))
```

```python
data_frame
```

```python
# Finding the average distance travelled by all the flights on friday which are delayed.
data_frame[(data_frame['arr_delay']>0) & (data_frame['weekday']==4)].mean()['distance']
```

```python
# Sum of total distance travelled by all the flights on friday which are delayed.
data_frame[(data_frame['arr_delay']>0) & (data_frame['weekday']==4)].sum()['distance']
```
